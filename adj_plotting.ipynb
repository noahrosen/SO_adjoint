```python
grid = xr.open_dataset('/glade/work/noahrose/emu_test/ECCO-GRID.nc')
```

```python
vname = 'qnet'

#fname = '/glade/work/noahrose/emu_samw/emu_adj_1_1_3_SO_total_mask.bin_1/output/o'
#fname = f'/glade/work/noahrose/emu_so_hc/emu_adj_1_1_3_SO_HC800_50S_mask.bin_1/output/adxx_{vname}.0000000129.data'
#fname = f'/glade/work/noahrose/emu_test/emu_adj_301_312_3_SO_HC2000_mask.bin_1/output/adxx_{vname}.0000000129.data'
fname = f'/glade/work/noahrose/emu_so_hc/emu_adj_301_312_3_SO_HC800_50S_mask.bin_1/output/adxx_{vname}.0000000129.data'

adjfile = np.fromfile(fname,dtype='>f4')

nx = 90
ny = 1170
nt = adjfile.shape[0]//nx//ny

adj_tiles = ecco.llc_compact_to_tiles(adjfile.reshape(nt,ny,nx))

# Regrid tile data to lat/lon using nearest neighbor

new_grid_delta_lat = 1
new_grid_delta_lon = 1

new_grid_min_lat = -90
new_grid_max_lat = 90

new_grid_min_lon = -180
new_grid_max_lon = 180

new_grid_lon_centers, new_grid_lat_centers,\
new_grid_lon_edges, new_grid_lat_edges,\
adj_rg =\
        ecco.resample_to_latlon(grid.XC, \
                                grid.YC, \
                                adj_tiles,\
                                new_grid_min_lat, new_grid_max_lat, new_grid_delta_lat,\
                                new_grid_min_lon, new_grid_max_lon, new_grid_delta_lon,\
                                fill_value = np.nan, \
                                mapping_method = 'nearest_neighbor',
                                radius_of_influence = 120000)
```

```python


lons = new_grid_lon_centers
lats = new_grid_lat_centers

fig,axs = plt.subplots(1,1,figsize=(8,4),subplot_kw={'projection':ccrs.PlateCarree()})

lag = 20*52


im1 = axs.pcolormesh(lons,lats,adj_rg[nt-lag-1],transform=ccrs.PlateCarree(),
                        cmap=cm.balance,
                        norm = colors.CenteredNorm())#,vmax = 0.00005,vmin=-0.00005)
eqline = axs.plot(np.arange(-180,180),np.zeros(360),'--k',transform=ccrs.PlateCarree())
maskline = axs.plot(np.arange(-180,180),np.zeros(360)-50,'--r',transform=ccrs.PlateCarree())
axs.add_feature(cfeature.LAND,facecolor='k',zorder=10)
plt.colorbar(im1,ax = axs, fraction=0.02, pad=0.04,
             label=f'Sensitivity to {vname}')
            #label=r'Sensitivity $(\frac{m}{N/m^2})$')
plt.title(f'Lag = {lag} weeks')
```

```python
adj_rg_norm = np.sqrt(np.sum((adj_rg/1e18)**2,axis=(1,2)))

plt.plot(np.arange(0,1300)/52,adj_rg_norm[1300:0:-1])
plt.xlabel('lag (years)')
plt.ylabel(f'Norm of sensitivity field to {vname}')
#plt.axvline(34,c='k')
```
