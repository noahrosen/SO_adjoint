```python
theta_ecco = xr.open_mfdataset('/glade/work/noahrose/ECCOV4r4/nctiles_monthly/THETA/*/*.nc').THETA
theta_ecco.load()
```

```python
basin_names = ['full','Atl','Pac','Ind']
mask_fnames = [f'SO_T700_{bname}_mask_v2.bin' for bname in basin_names]
nx = 90
ny = 1170
nz = 5265000//nx//ny
maskfiles_grid = [ecco.llc_compact_to_tiles(np.fromfile(mname,'>f4').reshape(nz,ny,nx)) 
                  for mname in mask_fnames]
```

```python
qoi_ts = [(theta_ecco * mask).sum(('i','j','k','tile')) for mask in maskfiles_grid]

qoi_ts = np.array([q.to_numpy() for q in qoi_ts])
```

```python
plt.plot(qoi_ts.T)
plt.legend(basin_names)
```

```python
theta_mean = theta_ecco.mean('time')
```

```python
ecco.plot_tiles(theta_mean.isel({'k':0}),layout='latlon',rotate_to_latlon=True,show_tile_labels=False,less_output=True,)
```

```python
nt = qoi_ts.shape[1]
qoi_ann = qoi_ts.reshape(4,nt//12,12)
```

```python
qoi_clim = qoi_ann.mean(axis=1)
qoi_anom = qoi_ts - np.tile(qoi_clim,nt//12)
```

```python
from scipy.signal import filtfilt,butter
b,a = butter(2,1/48,'lp')
qoi_filt = filtfilt(b,a,qoi_anom,axis=-1)
```

```python
plt.plot(theta_ecco.time,qoi_anom.T,ls='--',lw=0.5)
plt.gca().set_prop_cycle(None)
plt.plot(theta_ecco.time,qoi_filt.T)
plt.legend(basin_names)
plt.xlabel('time')
plt.ylabel(r'T$_{700}$ anomaly (K)')
```

```python
PmAI = qoi_ts[2] - (ratio_ind_atl * qoi_ts[3] + (1 - ratio_ind_atl) * qoi_ts[1])
```
